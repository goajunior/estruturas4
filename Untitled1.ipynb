# 2. Dimensionamento de Vigas de Aço à Flexão a Flexão Simples

O colapso de uma barra de aço submetido a esforços de flexão pode de diversos modos:

* formação de rótulas plásticas;

* instabilidade local dos elementos que compõem a seção transversal;

* flambagem lateral.




Tipos de aços estruturais para perfis:

|ASTM A36| 250 | 400 a 550| 1, 2 e 3|

Onde:

* Grupo 1 - perfis com espessura da mesa inferior ou igual a 37,5 $mm$.

* Grupo 2 - perfis com espessura da mesa superior a 37,5 $mm$ e inferior ou igual a 50 $mm$.

* Grupo 3 - perfis com espessura da mesa superior a 50 $mm$.

* Grupo 4 - perfis tubulares.


No dimensionamento de perfis como os perfis I e H, deve-se verificar a ocorrência de flambagem local da alma e mesa. Utilizaremos o conceito de índice de esbeltez:

$\lambda = \frac{b}{t}$.

Os gráficos que realcionam momento fletor com esbeltez (FLM e FLA, respectivamente) permitem classificar as seções em:

* supercompactas ($0<\lambda \leq \lambda_{pl}$)

* compactas ($\lambda_{pl} < \lambda \leq \lambda_{p}$)

* semicompactas ($\lambda_{p}<\lambda \leq \lambda_{r}$)

* esbeltas ($\lambda_{r}<\lambda $)

###  Exemplos



1. Determinar o $M_{rd}$ da seção transversal do perfil VS 600$\times$95 com aço ASTM A572 grau 50.

Dados: $b_f = 300 \; mm$,$t_f=12,5 \; mm$, $t_w=8\; mm$, $h = 575\; mm$, $W_x = 2580\times 10³ \;mm³$, $Z_x = 2860\times 10³ \;mm³$, aço com $f_y=345$ MPa

Verificação da mesa:

$\lambda_f = \frac{b_f}{2t_f}$

```python
lambdaf = 300/(2 * 12.5)
```

```python
lambdaf
```

$\lambda_p=0,38\sqrt{\frac{E}{f_y}}$

```python
import math
lambdap = 0.38*math.sqrt(200000/345)
```

```python
lambdap
```

$\lambda_f > \lambda_p$

$\lambda_r=0,95\sqrt{\frac{k_c\times E}{0,7 \times f_y}}$ (perfis soldados)

$k_c = \frac{4}{\sqrt{\frac{h}{t_w}}}$

$\lambda_w = \frac{h}{t_w}$

```python
lambdaw = 575/8
```

```python
lambdaw
```

```python
kc = 4/math.sqrt(lambdaw)
```

```python
kc
```

```python
lambdar = 0.95 * math.sqrt((kc * 200000)/(0.7 * 345))
```

```python
lambdar
```

$\lambda_p < \lambda_f < \lambda_r$ $\Rightarrow$ **mesa semicompacta**

Verificação da alma:

$\lambda_w = \frac{h}{t_w}$

```python
lambdaw
```

$\lambda_p =3,76\sqrt{\frac{E}{f_y}}$ (perfis soldados)


```python
lambdap = 3.76*math.sqrt(200000/345)
```

```python
lambdap
```

$\lambda_w < \lambda_p$

$M_r = 0,7 \times W_x \times f_y $ (**Momento de perda de proporcionalidade**)

$M_{pl} = Z_x \times f_y$ (**Momento de plastificação**)

```python
Mr = 0.7 * 2580e03 * 345
```

```python
Mr = Mr/1000/1000 #convertendo para KN m
Mr
```

```python
Mpl = 2860e03 * 345
```

```python
Mpl = Mpl/1000/1000
```

```python
Mpl
```

Momento resistente de cálculo:

$M_{rd} = 0,90 \times (M_{pl} - (M_{pl} - M_{r}) \times \frac{\lambda_f - \lambda_p}{\lambda_r - \lambda_p})$

```python
Mrd = 0.90 * ( Mpl - (Mpl - Mr) * (12 - 9.15)/(18.8 - 9.15))
Mrd
```

O momento resistente de cálculo será $791,4 KN\cdot m$.

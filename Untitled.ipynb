# Ações Atuantes em Uma Estrutura

As ações que atuam em uma estrutura são classificadas em três categorias:

* Permanentes (G): peso próprio da estrutura e peso de todos os componentes da construção: pisos, paredes, revestimentos e equipamentos fixos;


* Variáveis (Q): sobrecargas devido ao uso: divisórias, móveis, pressão hidrostática, empuxo de terra, vento e variação de temperatura;


* Excepcionais (E): ações de grande intensidade e baixa probabilidade de ocorrência: explosões, choques de veículoes e abalos sísmicos.

A norma NBR-8800/08 determina que a combinação de cargas deve obedecer ao seguinte somatório:

$ N = \sum (\gamma_{g} G) + E + \gamma_{q1} Q_1 + \sum_{j=2}^{n} (\gamma_{q_j} \psi_j Q_j)$

Para os coeficientes $\gamma_g$:

| Combinações | Peso próprio de estruuras metálicas | Peso de estruturas pré-moldadas| Peso próprio de estruturas moldadas no local | Peso próprio de elementos construtivos industrializados | Peso próprio de equipamentos|
|:-----------------:|:-----------------:|:-----------------:|:--------------:|:------------:|:--------------:|
|Normais| 1,25| 1,3| 1,35| 1,40| 1,50|
|Especiais| 1,15| 1,20| 1,25| 1,30| 1,40|
|Excepcionais| 1,10 | 1,15| 1,15| 1,20| 1,30|

**Obs: Quando a carga permanente atuar de uma maneira favorável para a resistênica da estrutura, admiti-se que $\gamma_g = 1,00$.**

Para os coeficientes $\gamma_q$:

| Combinações | Efeito de Temperatura | Ações de Vento | Demais ações devido ao uso ou ocupação | Ações truncadas|
|:-----------------:|:-----------------:|:-----------------:|:-------------:|:--------------:|
|Normais| 1,20| 1,4| 1,50|1,20| 
|Especiais| 1,00| 1,20| 1,30| 1,10|
|Excepcionais| 1,00 | 1,00| 1,00| 1,00|

Os coeficientes de redução $\psi_j$ são determinados pelas tabelas:

| Tipo da ação                             |         $\psi$| 
:---  |:--------------------------------------------------- |
|Locais sem ocorrência de pesos de equipamentos fixos e sem concentração de pessoas| 0,5|
Locais com ocorrência de pesos de equipamentos fixos ou com concentração de pessoas| 0,7| 
|Cargas acidentais para bibliotecas, arquivos, depósitos, oficinas, garagens e sobrecargas em coberturas| 0,8 |
|Pressão dinâmica do vento| 0,6|
|Variações de temperatura em relação à média anual| 0,6|
|Passarelas de pedestres| 0,6|
|Vigas de pontes| 1,0|
|Pilares de pontes| 0,7|



### Exemplo 1: Torre com Reservatório de Água



Dados:

* Capacidade do reservatório: $30.000 l$ (peso específico da água $= 10.000 N/m³$);

* Carga de vento nominal atuante no reservatório: $20 KN$;

* Carga permanente da torre e do reservatório: $48 kN$;

* A torre possui base quadrada de $6m \times 6m$.

Problema: calcular as ações de projeto nas colunas da torre.

#### Solução

Pela dupla simetria, temos que em duas colunas haverá esforços apenas de compressão e, nas outras duas colunas, haverá combinação de esforços de tração e compressão, além disso, devemos verificar uma situação em que o reservatório esteja completamente vazio (por que?).

Para as colunas comprimidas e considerando a carga do reservatório como a carga variável predominante:

$N_d=1,25\times(\frac{G}{4}) + 1,2\times (\frac{30\times 10}{4}) + 1,4\times 0,6\times(\frac{\frac{20 \times 14}{6}}{2})$

```python
Nd = 1.25*(48/4) + 1.2*(300/4) + 1.4*0.6*(20*14/6/2)
```

```python
Nd
```

Se tomarmos a carga devida ao vento como a ação predominante teremos a seguinte combinação:

$N_d=1,25\times(\frac{G}{4}) + 1,4\times(\frac{\frac{20 \times 14}{6}}{2}) + 1,2\times 0,8 \times (\frac{30\times 10}{4})$

```python
Nd = 1.25*(48/4) + 1.4*(20*14/6/2) + 1.2*0.8*(300/4)
```

```python
Nd
```

A carga de vento traciona os pilares da torre que estão à esquerda. Uma situação que deve ser considerada é quando o reservatório está vazio e adotando a carga devida ao vento como predominante:

$T_d = -1,0 \times (\frac{G}{4}) + 1,4 \times (\frac{\frac{20 \times 14}{6}}{2})$

```python
Td = -1.0*(48/4) + 1.4*(20*14/6/2)
```

```python
Td
```
